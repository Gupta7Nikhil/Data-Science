The grey background and general aesthetics of the Seaborn Python Library are well known. There are, however, not many other built-in styles available: darkgrid, white grid, dark, white and ticks. You can set themes using the set_style() function of seaborn library.

```python
# Data
data = np.random.normal(size=(20, 6)) + np.arange(6) / 2
 
# Proposed themes: darkgrid, whitegrid, dark, white, and ticks
 
sns.set_style("whitegrid")
sns.boxplot(data=data)
plt.title("whitegrid")
plt.show()
```

```python
sns.set_style("darkgrid")
sns.boxplot(data=data);
plt.title("darkgrid")
plt.show()
 
```

```python
sns.set_style("white")
sns.boxplot(data=data);
plt.title("white")
plt.show()

```

```python
sns.set_style("dark")
sns.boxplot(data=data);
plt.title("dark")
plt.show()

```

```python
sns.set_style("ticks")
sns.boxplot(data=data);
plt.title("ticks")
plt.show()
```

# Setting the size and color of the plot

```python
sns.set_style('whitegrid')
sns.lmplot(x ='total_bill', y ='tip', data = df, hue ='sex', 
           markers =['o', 'v'], scatter_kws ={'s':100}, 
           palette ='plasma')
```

Explanation
In this example what seaborn is doing is that its calling the matplotlib parameters indirectly to affect the scatter plots. We specify a parameter called scatter_kws. We must note that the scatter_kws parameter changes the size of only the scatter plots and not the regression lines. The regression lines remain untouched. We also use the palette parameter to change the color of the plot. Rest of the things remain the same as explained in the first example.

Explanation

```python
sns.jointplot(data=pdata, x="bill_length_mm", y="bill_depth_mm", hue = 'sex')
```

Additional keyword arguments can be passed down to the underlying plots:


```python

sns.jointplot(
    data=pdata, x="bill_length_mm", y="bill_depth_mm",
    marker="+", s=100, marginal_kws=dict(bins=25, fill=False),
)
```

# Use JointGrid parameters to control the size and layout of the figure:



```python
sns.jointplot(data=pdata, x="bill_length_mm", y="bill_depth_mm", height=5, ratio=2, marginal_ticks=True)
```

# To add more layers onto the plot, use the methods on the JointGrid object that jointplot() returns:


```python

g = sns.jointplot(data=pdata, x="bill_length_mm", y="bill_depth_mm")
g.plot_joint(sns.kdeplot, color="r", zorder=0, levels=6)
g.plot_marginals(sns.rugplot, color="r", height=-.15, clip_on=False)
```

```python
import pandas as pd
df = pd.read_csv ( "cars.csv" )
X1 = df [ [ 'Weight', 'Volume' ] ]
y = df [ 'CO2' ]
from sklearn import linear_model
regr = linear_model.LinearRegression()
regr.fit ( X1, y )

#predict the CO2 emission of a car where the weight is 2300kg, and the volume is 1300cm3:

predictedCO2 = regr.predict ( [ [ 2300, 1300 ] ] )
predictedCO2

#We have predicted that a car with 1.3 liter engine, and a weight of 2300 kg, will release approximately 
#107 grams of CO2 for every kilometer it drives.
```

```python
import pandas
from sklearn import linear_model

df = pandas.read_csv("cars.csv")

X = df[['Weight', 'Volume']]
y = df['CO2']

regr = linear_model.LinearRegression()
regr.fit(X, y)

print(regr.coef_)

#These values tell us that if the weight increase by 1kg, the CO2 emission increases by 0.00755095g.

#And if the engine size (Volume) increases by 1 cm3, the CO2 emission increases by 0.00780526 g.
```
